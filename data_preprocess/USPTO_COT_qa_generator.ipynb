```python
import json
import random

with open("cot_data_V2/cot_data_V2/uspto_cot_llava_train_V2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# with open("펍켐 json", "r", encoding="utf-8") as f:
#     data2 = json.load(f)

# 2중 리스트 풀기 (flatten)
def flatten(data):
    flat = []
    for item in data:
        if isinstance(item, list):  # 리스트 안에 리스트가 있을 때
            flat.extend(item)
        else:
            flat.append(item)
    return flat

data = flatten(data)
# data2 = flatten(data2)

# # 두 개 합치기
# merged = data1 + data2
# # 랜덤 셔플
# random.seed(42)
# random.shuffle(merged)

# # ID 재정렬
# for new_id, item in enumerate(merged):
#     item["id"] = f"{new_id:010d}"

# # 저장
with open("uspto_cot_llava_train_test.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
import json

with open("uspto_cot_llava_train_test.json", "r", encoding="utf-8") as f:
    data = json.load(f)

bad = []
for i, item in enumerate(data):
    if "image" not in item or not item["image"]:
        bad.append((i, item.get("id"), item.get("image")))
    if "conversations" not in item or not item["conversations"]:
        bad.append((i, item.get("id"), "missing conversations"))

print(f"총 {len(bad)}개 오류 항목 발견")
for b in bad[:10]:  # 앞부분만 출력
    print(b)
```
